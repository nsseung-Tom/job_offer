```python
def calculate_hourly_wage(df, hours_column, wage_column, unit_column, output_column):
    def calculate(row):
        hours_worked_weekly = row[hours_column]
        reservation_wage = row[wage_column]
        reservation_unit = row[unit_column]
        
        if hours_worked_weekly == 0:
            # Handle cases where hours_worked_weekly is zero to avoid ZeroDivisionError
            return None

        if reservation_unit in [1, 'Year', 'year']:  # Yearly or 'year'
            hourly_wage = reservation_wage / (hours_worked_weekly * 52)  # Assuming 52 weeks in a year
        elif reservation_unit in [2, 'Month', 'month']:  # Monthly
            hourly_wage = reservation_wage / (hours_worked_weekly * 4)  # Assuming 4 weeks in a month
        elif reservation_unit in [3, 'Week', 'week', ' Week'] or reservation_unit in ['every two weeks', 'Every two week']:  # Weekly or 'every two weeks'
            if reservation_unit in ['every two weeks', 'Every two week']:
                hours_worked_weekly *= 2  # Doubling the hours for 'every two weeks' scenario
            hourly_wage = reservation_wage / hours_worked_weekly
        elif reservation_unit == 4 or reservation_unit == 'Hour' or reservation_unit == 'hour':  # Hourly or 'hour'
            hourly_wage = reservation_wage
        else:
            # Handle unrecognized reservation_unit values
            return None

        return hourly_wage

    df[output_column] = df.apply(calculate, axis=1)
    return df
```

```python
# Previous_wage_hourly using Reservation Wage 
df_weekly_sum1 = calculate_hourly_wage(df_weekly_sum, 'how_many_hours_prefer_to_work_weekly', 'reservation_wage', 'reservation_unit', 'previous_wage_hourly')
df_weekly_sum2 = calculate_hourly_wage(df_weekly_sum1, 'weekly_working_hour', 'job_offer_wage','job_offer_unit', 'job_offer_wage_hourly')
df_weekly_sum3 = calculate_hourly_wage(df_weekly_sum2, 'accepted_weekly_working_hour', 'accepted_job_wage','accepted_job_unit', 'accepted_job_wage_hourly')
```

```python
# Assuming df_weekly_sum3 contains columns: acceptance_yn, offered_wage, previous_wage_hourly, accepted_job_wage_hourly
# and other basic columns like sex, race, age, caseid

# Create an initial final_df DataFrame with basic columns
basic_columns = ['caseid', 'received_job_offers', 'how_many_job_offers']
final_df = df_weekly_sum3[basic_columns].copy()

# Condition 1: df_weekly_sum3['acceptance_yn'] == 'Yes'
condition_1 = df_weekly_sum3['acceptance_yn'] == 'Yes'
final_df.loc[condition_1, 'offered_wage'] = df_weekly_sum3.loc[condition_1, 'job_offer_wage_hourly']
final_df.loc[condition_1, 'previous_wage'] = df_weekly_sum3.loc[condition_1, 'previous_wage_hourly']
final_df.loc[condition_1, 'acceptance_yn'] = 'y'

# Condition 3: df_weekly_sum3['acceptance_yn'] == 'No'
condition_3 = df_weekly_sum3['acceptance_yn'] == 'No'
final_df.loc[condition_3, 'offered_wage'] = df_weekly_sum3.loc[condition_3, 'job_offer_wage_hourly']
final_df.loc[condition_3, 'previous_wage'] = df_weekly_sum3.loc[condition_3, 'previous_wage_hourly']
final_df.loc[condition_3, 'acceptance_yn'] = 'n'
```
